## Matplotlib Workflow Example

```python
# Import libraries

%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]


# Setup plot

fig, ax = plt.subplots(figsize=(10, 10)) # (width, hegiht)

# Plot data

ax.plot(x, y)

# Customize plot

ax.set(title = "Simple Plot",
      xlabel = "X-Axis",
      ylabel = "Y-Axis")


# Save and show figure

fig.savefig("Images\sample-plot.png")
```

```python
# Make a plot from dictionary

nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 5,
                     "Cashew butter": 15}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax.set(title = "Butter Sales",
       xlabel = "Butter types",
       ylabel = "Number of jars");
```

```python
# Histogram

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two Options for Subplots

```python
# Option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10,5))


# Plot each axis

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(100));
```

```python
# Option 2

fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))

# Plot to each different index

ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(100));
```

### Which one should we use? (pyplot vs matplotlib object oriented method)

- Use pyplot to quickly plot
- Use matplotlib OO to plot something more advanced

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method

over_50.plot(kind= "scatter",
             x= "age",
             y= 'chol',
             c= 'target')
```

```python
##### OO method mixed with pyplot

fig, ax = plt.subplots(figsize=(10, 5))
over_50.plot(kind= "scatter",
             x= "age",
             y= 'chol',
             c= 'target',
             ax= ax);

ax.set_xlim([45, 100]);
```

```python
## OO method from scratch

fig,ax = plt.subplots(figsize=(10, 5))

scatter = ax.scatter(x=over_50["age"], y=over_50['chol'],
                     c=over_50["target"])

## Customize

ax.set(title= "Heart Disease and Cholesterol Levels",
       xlabel= "Age",
       ylabel= "Cholesterol");

## Add legend

ax.legend(*scatter.legend_elements(), title= "Target");

## Add horizontal line

ax.axhline(over_50["chol"].mean(),
           linestyle = '--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows= 2,
                               ncols= 1,
                               figsize= (10, 10),
                               sharex= True)

## Adding to ax0

scatter = ax0.scatter(x= over_50["age"],
                      y= over_50["chol"],
                      c= over_50["target"])

## Customize ax0

ax0.set(title= "Heart Disease and Cholesterol Levels",
        xlabel= "Age",
        ylabel= "Cholesterol");

# Add legends to ax0

ax0.legend(*scatter.legend_elements(), title= "Target")

## Add meanline

ax0.axhline(y= over_50["chol"].mean(),
            linestyle= "--");


## Adding to ax1

scatter_1 = ax1.scatter(x= over_50["age"],
                        y= over_50["thalach"],
                        c= over_50["target"])

# Customize ax1

ax1.set(title= "Heart Disease and Maximum Heart Rate",
        xlabel= "Age",
        ylabel= "Max Heart Rate");

## Add legends to ax1

ax1.legend(*scatter_1.legend_elements(), title= "Target")


## Add meanline

ax1.axhline(y= over_50["thalach"].mean(),
            linestyle= "--");

## Add title to figure

fig.suptitle("Heart Disease Analysis", fontsize= 16, fontweight= "bold");
```

```python
# Set style

plt.style.use('seaborn-whitegrid')

## OO method from scratch

fig,ax = plt.subplots(figsize=(10, 5))

scatter = ax.scatter(x=over_50["age"], 
                     y=over_50['chol'],
                     c=over_50["target"],
                     cmap= "summer") #color scheme

## Customize

ax.set(title= "Heart Disease and Cholesterol Levels",
       xlabel= "Age",
       ylabel= "Cholesterol");

## Add legend

ax.legend(*scatter.legend_elements(), title= "Target");

## Add horizontal line

ax.axhline(over_50["chol"].mean(),
           linestyle = '--');
```

```python
# Customizing x and y axis annotations

# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows= 2,
                               ncols= 1,
                               figsize= (10, 10),
                               sharex= True)

## Adding to ax0

scatter = ax0.scatter(x= over_50["age"],
                      y= over_50["chol"],
                      c= over_50["target"],
                      cmap= "summer")

## Customize ax0

ax0.set(title= "Heart Disease and Cholesterol Levels",
        xlabel= "Age",
        ylabel= "Cholesterol");

## Change ax0 x axis limits

ax0.set_xlim([50, 80])

# Add legends to ax0

ax0.legend(*scatter.legend_elements(), title= "Target")

## Add meanline

ax0.axhline(y= over_50["chol"].mean(),
            linestyle= "--");


## Adding to ax1

scatter_1 = ax1.scatter(x= over_50["age"],
                        y= over_50["thalach"],
                        c= over_50["target"],
                        cmap= "summer")

# Customize ax1

ax1.set(title= "Heart Disease and Maximum Heart Rate",
        xlabel= "Age",
        ylabel= "Max Heart Rate");

## Change ax1 x axis limits

ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

## Add legends to ax1

ax1.legend(*scatter_1.legend_elements(), title= "Target")


## Add meanline

ax1.axhline(y= over_50["thalach"].mean(),
            linestyle= "--");

## Add title to figure

fig.suptitle("Heart Disease Analysis", fontsize= 16, fontweight= "bold");
```
